```python
url = "http://web.mta.info/developers/turnstile.html"
res=requests.get(url)
page=BeautifulSoup(res.content)
mydivs = page.find_all("div", {"class": "span-84 last"})
links_all=[]
for div in mydivs:
    links=div.findAll('a')
links
for a in links:
    links_all.append("http://web.mta.info/developers/"+a['href'])
links_all

```
